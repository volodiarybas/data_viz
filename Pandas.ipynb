```python
 def ranking(x):
    if x['ind'] % 20 == 0:
        return 'champion'
    elif x['ind'] % 20 == 1:
        return 'Runner-up'
    elif (x['ind'] % 20 == 2) | (x['ind'] % 20 == 3):
        return 'Cl-zone'
    elif x['ind'] % 20 in [17,18,19]:
        return 'Relegation'
    else: return 'Average'
    
```

```python
pl['ind']=pl.index.values
pl
```

```python
pl.insert(0,'ranking',pl.apply(ranking, axis=1))
pl
```

# Task
## Build a plot with a line for each rank n=1 through n=3,
## where the line shows what fraction of that rank's roles
## were 'actor' roles for each year in the history of film.

```python
c = cast
c = c[c.n <= 3]
c = c.groupby(['year', 'type', 'n']).size()
c = c.unstack('type').fillna(0)
r = c.actor / (c.actor + c.actress)
r = r.unstack('n')
r.plot(ylim=[0,1])
```
